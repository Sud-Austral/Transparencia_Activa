```python
import pandas as pd
```

```python
dfDato19 = pd.read_csv(r"http://www.cplt.cl/transparencia_activa/datoabierto/archivos/TA_Nomina_beneficiarios.csv", encoding = "ISO-8859-1", sep=";")
dfDato19
```

```python
dfDato19.columns
```

```python
for i in dfDato19.columns:
    print(i)
    print(len(dfDato19[i].unique()))
```

```python
for i in dfDato19["nombre"].unique():
    print(i)
    print(len(dfDato19[dfDato19["nombre"] == i]))
```

```python
for i in dfDato19["Mes"].unique():
    #print(i)
    #print(len(dfDato19[dfDato19["Mes"] == i]))
    defAux = dfDato19[dfDato19["Mes"] == i]
    if(len(defAux) < 1000000):
        defAux.to_excel(f"TA_Nomina_beneficiarios_{i}.xlsx", index=False)
    else:
        defAux.iloc[:1000000,:].to_excel(f"TA_Nomina_beneficiarios_{i}_parte1.xlsx", index=False)
        defAux.iloc[1000000:,:].to_excel(f"TA_Nomina_beneficiarios_{i}_parte2.xlsx", index=False)
```
