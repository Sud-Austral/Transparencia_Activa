```python
import pandas as pd
```

```python
dfDato16 = pd.read_csv(r"http://www.cplt.cl/transparencia_activa/datoabierto/archivos/TA_PersonalContratohonorarios.csv", encoding = "ISO-8859-1", sep=";")  
```

```python
dfDato16["Año"] = dfDato16["fecha_ingreso"].apply(lambda x: str(x)[0:4])
```

```python
dfDato16Clean =dfDato16.drop(columns = ['fecha_publicacion', 'anyo','fecha_ingreso'])
```

```python
#dfGroup16 = dfDato16Clean.groupby(['organismo_nombre', 'organismo_codigo',  'Mes', 'denominacion', 
#           'Tipo Unidad monetaria', 'imputacion_presupuestaria', 'objeto_transferencia', 'razon_social',
#           'Nombre persona', 'apellido_paterno', 'apellido_materno', 'activado', 'Año']).sum()
dfGroup16 = dfDato16Clean.groupby(['camino', 'organismo_nombre', 'organismo_codigo', 'Mes', 'Nombres',
       'Paterno', 'Materno', 'grado_eus', 'descripcion_funcion',
       'tipo_calificacionp', 'region', 'Tipo Unidad monetaria',
       'pago_mensuales',
       'fecha_termino', 'observaciones', 'enlace', 'viaticos', 'activado',
       'Año']).sum()
dfGroup16 = dfGroup16.reset_index()
dfGroup16
```

```python
dfDato16.head(1).to_excel("Dato16prueba.xlsx")
```

```python
#dfPivot16 = dfGroup16.pivot( index=[ 'organismo_nombre', 'organismo_codigo', 'denominacion',
#           'Tipo Unidad monetaria', 
#            'imputacion_presupuestaria', 'objeto_transferencia', 'razon_social',
#           'Nombre persona', 'apellido_paterno', 'apellido_materno', 'activado','Año'], columns=['Mes'], values='monto')

dfPivot16 = dfGroup16.pivot( index=[ 'organismo_nombre', 'organismo_codigo','Tipo Unidad monetaria',
                                    'pago_mensuales','Nombres','Paterno', 'Materno', 'activado', 'Año'],
                            columns=['Mes'], values='monto')
dfPivot16 = dfPivot16.reset_index()
```

```python
dfPivot16["Año"].unique()
```

```python
for i in dfPivot16["Año"].unique():
    dfAux = dfPivot16[dfPivot16["Año"] == i]
    dfAux.to_excel(f"TA_PersonalContratohonorarios/TA_PersonalContratohonorarios_{i}.xlsx", index=False)
```
