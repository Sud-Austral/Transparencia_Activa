```python
import pandas as pd
```

```python
pch = pd.read_csv(r"http://www.cplt.cl/transparencia_activa/datoabierto/archivos/TA_PersonalContratohonorarios.csv", encoding = "ISO-8859-1", sep=";")
pch
```

```python
pch["Año"] = pch["fecha_ingreso"].apply(lambda x: str(x)[0:4])
pch
```

```python
pch["Año"].unique()
```

```python
pchClean =pch.drop(columns = ['fecha_publicacion', 'anyo','fecha_ingreso', 'camino'])
pchClean
```

```python
for i in pchClean["Año"].unique():
    print(i)
    print(len(pchClean[pchClean["Año"]==i]))
```

```python
pchClean.columns
```

```python
#'pago_mensuales', 'viaticos','remuneracionbruta','remuliquida_mensual',
pchGroup = pchClean.groupby(['organismo_nombre', 'organismo_codigo', 'Mes', 'Nombres', 'Paterno',
       'Materno', 'grado_eus', 'descripcion_funcion', 'tipo_calificacionp',
       'region', 'Tipo Unidad monetaria',
        'fecha_termino','observaciones', 'enlace', 'activado', 'Año']).sum()
pchGroup = pchGroup.reset_index()
pchGroup
```

```python
pchGroup.columns
```

```python


dfPivot16 = pchGroup.pivot( index=[ 'organismo_nombre', 'organismo_codigo', 'Nombres', 'Paterno',
       'Materno', 'grado_eus', 'descripcion_funcion', 'tipo_calificacionp',
       'region', 'Tipo Unidad monetaria', 'fecha_termino', 'observaciones',
       'enlace', 'activado', 'Año'], 
                            columns=['Mes'], values=['remuneracionbruta', 'remuliquida_mensual','viaticos'])
dfPivot16 = dfPivot16.reset_index()
dfPivot16
```
