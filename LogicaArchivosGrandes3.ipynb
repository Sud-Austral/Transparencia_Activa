```python
import pandas as pd
```

```python
dfDato14 = pd.read_csv(r"http://www.cplt.cl/transparencia_activa/datoabierto/archivos/TA_PersonalContrata.csv", encoding = "ISO-8859-1", sep=";")
```

```python
dfDato14["Año"] = dfDato14["fecha_ingreso"].apply(lambda x: str(x)[6:10])
```

```python
dfDato14["camino"].unique()
```

```python
dfDato14Clean =dfDato14.drop(columns = ['fecha_publicacion', 'anyo','fecha_ingreso'])
```

```python
for i in dfDato14Clean["Año"].unique():
    print(i)
    print(len(dfDato14Clean[dfDato14Clean["Año"]==i]))
```

```python
dfGroup14 = dfDato14Clean.groupby(['organismo_nombre', 'organismo_codigo',  'Mes', 'denominacion', 
           'Tipo Unidad monetaria', 'imputacion_presupuestaria', 'objeto_transferencia', 'razon_social',
           'Nombre persona', 'apellido_paterno', 'apellido_materno', 'activado', 'Año']).sum()
dfGroup14 = dfGroup14.reset_index()
```

```python
dfPivot14 = dfGroup14.pivot( index=[ 'organismo_nombre', 'organismo_codigo', 'denominacion',
           'Tipo Unidad monetaria', 
            'imputacion_presupuestaria', 'objeto_transferencia', 'razon_social',
           'Nombre persona', 'apellido_paterno', 'apellido_materno', 'activado','Año'], columns=['Mes'], values='monto')
dfPivot14 = dfPivot14.reset_index()
```

```python
for i in dfPivot14["Año"].unique():
    dfAux = dfPivot14[dfPivot14["Año"] == i]
    dfAux.to_excel(f"TA_PersonalContrata/TA_PersonalContrata_{i}.xlsx", index=False)
```
